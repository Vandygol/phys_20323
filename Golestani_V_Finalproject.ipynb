```python
import matplotlib.pyplot as plt
import numpy as np

# Decay constants and half-lives
half_life_Rn_222 = 3.82 * 24 * 60  # Convert days to minutes
half_life_Po_218 = 3.1
half_life_Pb_214 = 26.8
half_life_Bi_214 = 19.9
half_life_Tl_207 = 1.3

# Initial number of atoms
initial_atoms = 20000
time_points = np.arange(0, 18000, 0.1) 

# Decay equations
decay_Rn_222 = initial_atoms * np.exp(-time_points / half_life_Rn_222)
decay_Po_218 = (initial_atoms - decay_Rn_222) * np.exp(-time_points / half_life_Po_218)
decay_Pb_214 = (initial_atoms - decay_Rn_222 - decay_Po_218) * np.exp(-time_points / half_life_Pb_214)
decay_Bi_214 = (initial_atoms - decay_Rn_222 - decay_Po_218 - decay_Pb_214) * np.exp(-time_points / half_life_Bi_214)
decay_Tl_207 = (initial_atoms - decay_Rn_222 - decay_Po_218 - decay_Pb_214 - decay_Bi_214) * np.exp(
    -time_points / half_life_Tl_207)
decay_Bp_207 = (initial_atoms - decay_Rn_222 - decay_Po_218 - decay_Pb_214 - decay_Bi_214 - decay_Tl_207)

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(time_points, decay_Rn_222, label='Rn-222')
plt.plot(time_points, decay_Po_218, label='Po-218')
plt.plot(time_points, decay_Pb_214, label='Pb-214')
plt.plot(time_points, decay_Bi_214, label='Bi-214')
plt.plot(time_points, decay_Tl_207, label='Tl-207')
plt.plot(time_points, decay_Bp_207, label='Bp-207')

plt.title('Radioactive Decay Chain')
plt.xlabel('Time (minutes)')
plt.ylabel('Number of Atoms')
plt.legend()
plt.grid(True)
plt.show()
```
